```
def input_fn(features, labels, training=True, batch_size=256):
    """An input function for training or evaluating"""
    dataset = tf.data.Dataset.from_tensor_slices((dict(features), labels))
    print('The dataset is:')
    print(dataset)
    print("__________________________________").
    if training:
        dataset = dataset.shuffle(1000).repeat()

    return dataset.batch(batch_size)
my_feature_columns = []
for key in DfTrain.keys():
    my_feature_columns.append(tf.feature_column.numeric_column(key=key))
#input_fn(DfTrain,y_train) 
#print("This is the feature column:\n" + str(my_feature_columns) +"\n"+ "This is my label column:'n" +str(y_train) )



classifier = tf.estimator.DNNClassifier(
    feature_columns=my_feature_columns,

    hidden_units=[30,30,90,180,360,720,1800,2],

    n_classes=2,
    model_dir = "/content/drive/MyDrive/Colab Notebooks/ML CONFLICT ANALYSIS/Data folder/Model1",)

classifier.train(
    input_fn=lambda: input_fn(DfTrain, y_train, training=True),
    steps=50000)

eval_result = classifier.evaluate(
    input_fn=lambda: input_fn(DfEval, y_eval, training=False))

print('\nTest set accuracy: {accuracy:0.3f}\n'.format(**eval_result))


def input_fn(features, batch_size=256):
    """An input function for prediction."""
    # Convert the inputs to a Dataset without labels.
    return tf.data.Dataset.from_tensor_slices(dict(features)).batch(batch_size)

predictions = classifier.predict(
    input_fn=lambda: input_fn(DfEval))

for pred_dict, expec in zip(predictions, y_eval): #fix ME!!!
    class_id = pred_dict['class_ids'][0]
    probability = pred_dict['probabilities'][class_id]
    print('\nPrediction is "{}" ({:.1f}%), expected "{}"'.format(y_eval[class_id], 100 * probability, expec))

```
